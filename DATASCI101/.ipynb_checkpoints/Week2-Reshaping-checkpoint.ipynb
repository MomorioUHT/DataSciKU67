```python
import pandas as pd

player_df = pd.read_csv('players.csv').loc[:, :'saves']
player_df
```

```python
#use Surname as index
player_df.set_index('surname')
```

```python
multi_index_player = player_df.set_index(['team', 'surname'])
multi_index_player
```

```python
multi_index_player.loc[('Algeria', 'Abdoun'), :]
```

```python
#Create MultiIndex to assign to columns
x = pd.MultiIndex.from_tuples( [('Info', 'position'),
                           ('Info', 'minutes'),
                           ('Attack', 'shots'),
                           ('Attack', 'passes'),
                           ('Defense', 'tackles'),
                           ('Defense', 'saves')])
```

```python
multi_index_player.columns = x
multi_index_player
```

```python
multi_index_player.loc['Algeria', 'Attack']
```

```python
#นอกสุด = level 0
multi_index_player.xs('Torres', level=1)

#ดูสถิติของ Torres มี Torres จาก 3 ทีม
```

```python
multi_index_player.xs('Algeria', level=0)
```

```python
#Sorting index

multi_index_player.sort_index(ascending=[True,False]) #Sort ตามชั้น
```

```python
#Swaplevel
multi_index_player.swaplevel() #Swap surname กับ team (ในกับนอก)
```

```python
#What if having multiple levels????
df_3_lvl = player_df.set_index(['team','position','surname'])
df_3_lvl
```

```python
df_3_lvl.swaplevel(0,2) #0 team, 2 surname
```

```python
df_3_lvl.swaplevel(1,0).sort_index()
```

```python
#use () as a frame to shorthand
(
    pd.read_csv('players.csv').loc[:, :'saves']
                              .set_index(['team','position','surname'])
                              .swaplevel(1,0)
                              .sort_index()
)
```

```python
## Pivot And Melt

#ข้ามไป 1 แถว คือ Avg temperatures และ slice ถึงแค่ dec
eu_temp = pd.read_csv('eu_city_temp.csv', skiprows=1).loc[:, :'Dec']
eu_temp

#อยากได้ Country, City, Month, Temperature
          #Au    Vienna  Jan    -0.5\n(32.5)
          #Au    Vienna  Feb    1.5\n(34.7)
          #...
```

```python
eu_temp.melt(id_vars=['Country', 'City']) #ระบุ Column ที่เราจะยุบรวม หรือคงที่ไว้
```

```python
#ระบุชื่อ var, value ให้
long_eu_temp = eu_temp.melt(id_vars=['Country', 'City'], 
                            var_name='Month', 
                            value_name='Temperature') 

long_eu_temp
```

```python
long_eu_temp.pivot(index='City', columns='Month', values='Temperature')
#กลับมาเป็นตารางแบบ Wide ใหม่
#สังเกตว่าคำสั่ง pivot จะ sort index ให้ด้วย!
```

```python
long_eu_temp.pivot(index=['Country','City'], 
                   columns='Month', 
                   values='Temperature')
#Multiindex
```

```python
player_df.groupby('team').passes.mean()
```

```python
#pivot table

#                forward          defender
#Algeria                sum of passes
#Argentina

player_df.pivot_table(index='team', 
                      columns='position', 
                      values='passes', 
                      aggfunc='sum')
```

```python
mul_g = player_df.groupby(['team','position']).passes.sum()
mul_g
```

```python
mul_g.to_frame()
```

```python
mul_g.to_frame().unstack() #Unstack เอาแถวชั้นในสุดเป็นคอลัมน์ชั้นในสุด
```
