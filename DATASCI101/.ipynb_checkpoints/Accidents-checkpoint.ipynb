```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
df = pd.read_csv('Accidents.csv')
df.head()
```

# mean population

```python
df.อายุ.mean() #Population
```

```python
#สุ่มกลุ่มตัวอย่าง 10 ตัว
s1 = df.sample(10)
#s1
s1.อายุ.mean()
```

# ลองหา mean อายุดูทีละ 10 คน 50 รอบแบบสุ่ม

```python
ls = []
for i in range(50):
    ls.append(df.sample(10).อายุ.mean())
```

```python
#ทำเป็น Series
s_10 = pd.Series(ls)
s_10.sort_values() 
```

### สังเกตว่าค่าค่อนข้างแกว่งไปแกว่งมาจาก mean ของ Population เยอะ

# ลองเพิ่มขนาดของกลุ่มตัวอย่าง เอาสัก 50 ตัว

```python
ls2 = []
for i in range(50):
    ls2.append(df.sample(50).อายุ.mean())
```

```python
#ทำเป็น Series
s_50 = pd.Series(ls2)
s_50.sort_values() 
```

### ดีขึ้น

# 100 ตัวเลย!

```python
ls3 = []
for i in range(50):
    ls3.append(df.sample(100).อายุ.mean())
```

```python
#ทำเป็น Series
s_100 = pd.Series(ls3)
s_100.sort_values() 
```

### ยิ่งใกล้เคียง mean population เข้าไปอีก

# ลองหา Varianit

```python
s_10.var(), s_50.var(), s_100.var()
```

### ดังนั้นข้อสรุปคือ ยิ่งกลุ่มตัวอย่างที่ Sample ออกมา ยิ่งเยอะเท่าไหร่ ค่า mean ก็จะยิ่งใกล้กับ Population มากขึ้น!

```python
s_100.var(ddof=1), s_100.var(ddof=0)
```

```python
df.อายุ.var(ddof=1), df.อายุ.var(ddof=0)
```

```python
s_10.var(ddof=1), s_10.var(ddof=0)
```

# meaning

```python
resamples  = pd.DataFrame()

for i in range(30):
    one_sample = df.sample(50).อายุ
    
    ls = []
    for j in range(50):
        ls.append(one_sample.sample(50, replace=True).mean())
    resamples = pd.concat([resamples, pd.Series(ls)], axis=1)

resamples
```

```python
resamples.quantile([0.95, 0.05])
```
