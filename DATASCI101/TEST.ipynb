## Filtering

```python
player_df = player_df.reset_index()
player_df
```

```python
player_df.loc[ player_df.shots >= 20, :]
```

```python
player_df.loc[ 300 <= player_df.passes] #short form ที่ไม่ต้องระบุ column
```

```python
player_df.loc[ player_df.team == 'England']
```

```python
player_df[ player_df.team.str.endswith("land")] #ต้องเรียก string method ผ่าน dot str ก่อน!!
```

```python
player_df[player_df.surname.str.contains('rr')] #NOT REGEX!
```

```python
player_df[ (player_df.team == 'Spain') & (player_df.passes >= 300)] #NOT 'and' uses '&'
```

```python
player_df[ (player_df.team == 'Spain') | (player_df.passes >= 300)] #NOT 'or' uses '|'
```

```python
player_df.query('team == "Spain" and passes > 300') #สั้นกว่า
```

```python
df2 = player_df.sample(5) #without seed, amount = 5
df2
```

```python
df2 = player_df.sample(5, random_state=1334) #seed 1334, amount = 5
df2
```

```python
df2.passes + df3.passes #pandas จะบวกตามคู่เอง ตาม index ตรงกัน!
```

```python
df2.passes.add(df3.passes, fill_value=0) #ถ้ามี NaN ให้เอา 0 ไปบวกแทน
```
