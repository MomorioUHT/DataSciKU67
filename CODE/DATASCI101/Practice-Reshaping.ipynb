```python
import pandas as pd
```

```python
eu_temp = pd.read_csv("eu_city_temp.csv", skiprows=1).loc[:,:'Dec']
eu_temp
```

```python
#Make Long df
eu_temp_long = pd.melt(eu_temp, id_vars=['Country','City'], var_name='Month', value_name='Temperature')
eu_temp_long
```

```python
#Long to Wide 2
eu_temp_wide_2 = eu_temp_long.pivot(index='City', columns='Month', values='Temperature')
eu_temp_wide_2
```

```python
#Multi index 
eu_temp_multi = eu_temp_long.pivot(
    index=['Country', 'City'],
    columns='Month',
    values='Temperature'
)
eu_temp_multi
```

```python
# Sample DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concatenate along rows (axis=0)
result = pd.concat([df1, df2])
result = pd.concat([df1, df2], axis=1)
print(result)
```

```python
#Multi index  2
eu_temp_multi_2 = eu_temp_long.pivot_table(
    index=['Country', 'City'],
    columns='Month',
    values='Temperature',
    aggfunc='sum'
)
eu_temp_multi_2
```
