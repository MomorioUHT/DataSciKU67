# 6

```python
medals = medals.merge(
    drinks[["country", "continent"]],
    how="left",
    left_on="Country",
    right_on="country"
).drop(columns=["country"]) 

medals
```

# 7

```python
countries_with_nan_continent = medals.loc[medals["continent"].isnull(), "Country"]
countries_with_nan_continent
```

# 8

```python
def set_continent(country):
    if country in ["Australasia"]:
        return "OC"
    elif country in ["Canada", "United States", "Mexico"]:
        return "NA"
    elif country in ["Cuba", "Jamaica"]:
        return "SA"
    elif country in ["Czech", "Yugoslavia", "Germany", "Great Britain", "Russia", "Soviet"]:
        return "EU"
        
    return "OT"

medals["continent"] = medals["continent"].fillna(medals["Country"].apply(set_continent))
medals

# Check NaN Again
# medals["continent"].isnull().sum()
```

# 9

```python
medals = medals.set_index(["continent", "Country"]).drop("Code", axis=1)
medals
```

# 10

```python
medals = medals.sort_index(ascending=[False, True])

# use in future
medals_without_stack = medals.copy()

medals
```

# 11

```python
x = pd.MultiIndex.from_tuples([("Summer", "SummerGame"),
                              ("Summer", "SummerGold"),
                              ("Summer", "SummerSilver"),
                              ("Summer", "SummerBronze"),
                              ("Winter", "WinterGame"),
                              ("Winter", "WinterGold"),
                              ("Winter", "WinterSilver"),
                              ("Winter", "WinterBronze")])
medals.columns = x
medals
```

# 12

```python
medals = medals.swaplevel(axis=1)
medals 
```

# 13

```python
medals_without_index = medals_without_stack.reset_index()

medals_long = pd.melt(
    medals_without_index,
    id_vars=['continent', 'Country', 'SummerGame', 'WinterGame'], 
    value_vars=['SummerGold', 'SummerSilver', 'SummerBronze', 'WinterGold', 'WinterSilver', 'WinterBronze'],
    var_name='Medal',  
    value_name='Count'
)

medals_long['Season'] = medals_long['Medal'].apply(lambda x: 'Summer' if 'Summer' in x else 'Winter')
medals_long['Medal'] = medals_long['Medal'].apply(lambda x: x.replace('Summer', '').replace('Winter', ''))

medals_long['Game'] = medals_long.apply(
    lambda row: row['SummerGame'] if row['Season'] == 'Summer' else row['WinterGame'], axis=1
)

medals_long = medals_long.drop(columns=['SummerGame', 'WinterGame'])
medals_long = medals_long[['continent', 'Country', 'Season', 'Game', 'Medal', 'Count']]

medals_long
```

# 17

```python
#From long table
summary_table = medals_long.pivot_table(
    index='Medal',  
    columns='Season', 
    values='Count', 
    aggfunc='sum', 
    fill_value=0
)

summary_table
```

# 18

```python
summer_gold_data = medals_long[(medals_long['Season'] == 'Summer') & (medals_long['Medal'] == 'Gold')]

average_gold_per_country = summer_gold_data.groupby('continent')['Count'].mean()
average_gold_per_country
```

# 19

```python
#from long table

summer_data = medals_long[medals_long['Season'] == 'Summer']
winter_data = medals_long[medals_long['Season'] == 'Winter']

summer_count = summer_data.groupby(['continent'])['Country'].nunique()
winter_count = winter_data.groupby(['continent'])['Country'].nunique()

olympic_participation = pd.DataFrame({
    'Summer': summer_count,
    'Winter': winter_count
})

olympic_participation
```
